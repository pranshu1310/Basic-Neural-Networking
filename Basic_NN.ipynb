# Basic NN using the loss function as L1-norm or L2-norm:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 1000

xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(-10, 10, (observations,1))

inputs = np.column_stack((xs,zs))
print (inputs.shape)
```

```python
noise = np.random.uniform(-1, 1, (observations,1))

targets = 2*xs - 3*zs + 5 + noise

print (targets.shape)
```

```python
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, zs, targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=400)
plt.show()
# the azim parameter is used for plotting the data from different angles
```

```python
targets = targets.reshape(observations,1)
```

```python
init_range = 0.1
weights = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))
biases = np.random.uniform(low=-init_range, high=init_range, size=1)

print ("The weights are ",weights[0]," and ", weights[1])
print ("The bias is ",biases)
#the number of biases depend on the number of target variables. Here, there is only one target variable and hence one bias. 
```

```python
# Different learning rates affect the speed of optimization

learning_rate = 0.02
```

```python
for i in range (100):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
        
    loss = np.sum(deltas ** 2) / 2 / observations
    
    print (loss)
  
    deltas_scaled = deltas / observations
    
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print (weights, biases)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

# 5. **Тry with the L1-norm loss, given by the sum of the ABSOLUTE value of yj - tj. The L1-norm loss is given by:**
## $$ \Sigma_i = |y_i-t_i| $$

```python
observations = 1000

xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(-10, 10, (observations,1))

inputs = np.column_stack((xs,zs))

print (inputs.shape)
```

```python
noise = np.random.uniform(-1, 1, (observations,1))

targets = 2*xs - 3*zs + 5 + noise

print (targets.shape)
```

```python
targets = targets.reshape(observations,)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot(xs, zs, targets)

ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')

ax.view_init(azim=100)

plt.show()

targets = targets.reshape(observations,1)
```

```python
init_range = 0.1
weights = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))
biases = np.random.uniform(low=-init_range, high=init_range, size=1)
print (weights)
print (biases)
```

```python
learning_rate = 0.02
```

```python
for i in range (100):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
    loss = np.sum(np.absolute(deltas)) / 2 / observations
    print (loss)
    deltas_scaled = deltas / observations

    weights = weights - learning_rate * 2 * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * 2 * np.sum(deltas_scaled)
```

```python
print (weights, biases)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

# **Solution:**

1. Any function that has the property to be lower for better results and higher for worse results can be a loss function.
This includes the L1-norm loss.
2. Almost everything seems identical.
3. The values of the loss we generally lower (as we don't elevate to the second power).
4. For our problem, both L1-norm and L2-norm work equally well. 
5. Generally, L1-norm has more limitations than L2-norm. Overall, L2-norm loss is much more stable and computationally efficient.
